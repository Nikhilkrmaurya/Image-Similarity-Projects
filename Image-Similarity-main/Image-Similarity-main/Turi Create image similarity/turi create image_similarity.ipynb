```python
import turicreate as tc
```

```python
from google.colab import drive

drive.mount('/content/gdrive')
```

```python
# Load images from the downloaded data

reference_data  = tc.image_analysis.load_images('/content/gdrive/MyDrive/Colab Notebooks/Training/glio')
reference_data = reference_data.add_row_number()
```

```python
reference_data.explore()
```

```python
model = tc.image_similarity.create(reference_data)
```

```python
reference_data[:25]
```

```python
query_results = model.query(reference_data[0:3000], k= 7)
query_results.head()
```

```python
reference_data[2025]["image"].show()
```

```python
similar_rows = query_results[query_results['query_label'] == 2025]['reference_label']
reference_data.filter_by(similar_rows, 'id').explore()
```
